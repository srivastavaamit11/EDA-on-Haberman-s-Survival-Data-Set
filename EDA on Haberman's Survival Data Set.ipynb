# EDA on Haberman's Survival Data Set

This dataset has been taken from Kaggle which contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of the patients who had gone through surgery of breast cancer.

It contains four attributes 
1) Age : age of the patient
2) Year : the year in which the patient has gone through surgery
3) Axil_node : Number of positive Axilliary Nodes detected 
4) Status : Survival Status (classification attribute)
    1 = the patient survived 5 years or longer
    2 = the patient died within 5 year

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
hb=pd.read_csv("./haberman.csv")
hb.shape
hb.columns
```

```python
hb.plot(kind="scatter", x="age" , y="status")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hb,hue="status",height=4).map_dataframe(sns.scatterplot,x="age",y="year").add_legend()  #,hue="status"
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hb,hue="status",height=4).map(plt.scatter,"age","axil_node").add_legend()
plt.show()
```

```python
plt.close()
```

```python
sns.set_style("whitegrid")
sns.pairplot(hb,hue="status",height=5).add_legend()
#plt.show()
```

```python
sns.FacetGrid(hb,hue="status",height=5).map(sns.histplot,"age").add_legend()
```

```python
sns.violinplot(x="status", y="axil_node", data=hb, size=8)
plt.show()
```
